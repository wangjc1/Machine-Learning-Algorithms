```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
```

```python
#cumsum对前面几项累加后得到一个新的队列
Series([1,2,3,4,5]).cumsum()
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
#画柱图，点数不能太多，所以只取10个点
s1[:10].plot(kind='bar')
```

```python
#plot默认kind='line'，默认画的是线条，,grid=True是否显示方格
s1.plot(kind='line',grid=True, label='S1', title='This is Series')
s2.plot(label='S2',grid=True)
```

```python
plt.legend()
plt.show()
```

### 使用subplots的plot函数画图

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(s1)
ax[1].plot(s2)
plt.show()
```

### 使用Series的plot函数画图

```python
fig, ax = plt.subplots(2,1)
s1[0:10].plot(ax=ax[0], label='S1', kind='bar')
s2[0:10].plot(ax=ax[1], label='S2')
plt.show()
```

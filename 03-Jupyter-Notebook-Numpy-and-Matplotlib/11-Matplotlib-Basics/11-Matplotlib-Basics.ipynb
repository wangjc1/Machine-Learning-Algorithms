### 11 ``matplotlib`` 基础

```python
import numpy as np 
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot(x, y) #以x为横轴，y为纵轴的图形
```

```python
plt.show()
```

```python
siny = y.copy()
```

```python
cosy = np.cos(x)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="yellow")
plt.show()
```

关于color参数：[https://matplotlib.org/2.0.2/api/colors_api.html](https://matplotlib.org/2.0.2/api/colors_api.html)

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.show()
```

关于linestyle参数：[https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html](https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html)

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15) #指定X坐标区间在 (-5，15) 之间
plt.ylim(0, 1) #指定Y坐标区间在 (0, 1) 之间
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2]) #调节X范围为（-1, 11），Y范围为（ -2, 2）
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlabel("x axis") #x标签
plt.ylabel("y value")  #y标签
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend() #图示
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to matplotlib world!")
plt.show()
```

### 将两条曲线写到一个plot函数中，r--表示画红色的线

```python
plt.plot(x, siny,'r--',x, cosy,'y--')
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
```

```python
plt.axis([-10, 10,-10, 10]) 
#当x、y都是数组时，x所有的值的都是横坐标上的点
plt.plot([0,2,8,-4,-8], [0,2,8,-4,-8], color="red")
```

### Scatter Plot 散点图

```python
plt.scatter(x, siny, color="red")
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)

plt.scatter(x, y, alpha=0.5) #alpha是透明度
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
numpy.linspace(start, stop[, num=50[, endpoint=True[, retstep=False[, dtype=None]]]]])
返回在指定范围内的均匀间隔的数字（组成的数组），也即返回一个等差数列
start - 起始点，
stop - 结束点
num - 元素个数，默认为50，
"""
x = np.linspace(0.0, 5.0)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
```

```python
plt.plot(x, y1, 'b--', label='sin(pi*x)')
plt.ylabel('y1 value')
plt.plot(x, y2, 'r--', label='sin(pi*2x)')
plt.ylabel('y2 value')
plt.xlabel('x value')
plt.title('this is x-y value')
plt.legend()
```

```python
plt.show()
```

```python
#两行两列的第一个
plt.subplot(221)
plt.plot(x, y1, 'b--')
plt.ylabel('y1')

plt.subplot(222)
plt.plot(x, y2, 'r--')
plt.ylabel('y2')
plt.xlabel('x')

plt.subplot(223)
plt.plot(x, y1, 'r*')

plt.subplot(224)
plt.plot(x, y1, 'b*')
```

```python
plt.show()
```

### 用subplots也可以画多行多列图

```python
figure, ax = plt.subplots(2,2)
#第一行第一个图
ax[0][0].plot(x, y1)
#第一行第二个图
ax[0][1].plot(x, y2)
plt.show()
```

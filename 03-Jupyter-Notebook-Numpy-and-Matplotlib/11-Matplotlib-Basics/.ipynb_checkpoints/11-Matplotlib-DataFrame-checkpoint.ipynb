## 对于矩阵数据画图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
df = DataFrame(
    np.random.randint(1,10,40).reshape(10,4),
    columns=['A','B','C','D']
)
df.head()
```

### 按列画图

```python
df.plot()
plt.show()
```

```python
df.plot(kind='bar')
plt.show()
```

```python
df.plot(kind='barh')
plt.show()
```

```python
#stacked=True表示堆叠显示，把同一索引的数据显示到一个柱状里
df.plot(kind='bar', stacked=True)
plt.show()
```

```python
#本质上是一个线性图，然后用不同颜色填充了线条和轴之间的面积
df.plot(kind='area')
plt.show()
```

### 按行画图

```python
#取df中的第5行数据
a = df.iloc[5]
a
```

```python
a.plot()
plt.show()
```

```python
#遍历df中的所有行，用每一行的数据画线条
for i in df.index:
    df.iloc[i].plot(label=str(i))
plt.legend()
plt.show()
```

```python
#用一列数据画图
df['A'].plot()
plt.show()
```

```python
df.T.plot()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. 背景分格设置：axes_style and set_style

```python
%matplotlib inline
```

```python
x = np.linspace(0,14,100)
```

```python
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sinplot():
    plt.plot(x, y1)
    plt.plot(x, y2)
```

```python
sinplot()
```

```python
import seaborn as sns
```

```python
sinplot()
```

### 定义5种背景主题

```python
style = ['darkgrid', 'dark', 'white','whitegrid', 'tricks']
```

```python
sns.set_style(style[0], {'grid.color': 'red'})
```

```python
sinplot()
```

```python
sns.axes_style()
```

```python
sinplot()
```

### 重置默认主题

```python
sns.set()
```

```python
sinplot()
```

```python
#设置系统主题
sns.set_style('white')
sinplot()
```

```python
sns.set()
```

```python
sinplot()
```

# 2. 设置线条主题：plotting_context() and set_context() 

```python
context = ['paper', 'notebook', 'talk', 'poster']
```

```python
#可以设置网格线的宽度，线条的宽度等属性
sns.set_context(context[2], rc={'grid.linewidth': 1.0,'lines.linewidth':5})
```

```python
sinplot()
```

```python
sns.set()
```

```python
sns.plotting_context()
```

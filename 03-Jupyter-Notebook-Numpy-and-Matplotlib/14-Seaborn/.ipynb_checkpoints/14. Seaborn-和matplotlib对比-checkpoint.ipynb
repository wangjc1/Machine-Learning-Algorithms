```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets 
from pandas import Series, DataFrame
%matplotlib inline
```

## 加载鸢尾花数据集

```python
iris = datasets.load_iris()
#print(iris.DESCR)
```

```python
data = iris.data
data[:5]
```

```python
iris.target
```

```python
iris.target_names
```

```python
names = [iris.target_names[y] for y in iris.target ]
len(names)
```

```python
iris_new = DataFrame(iris.data)
iris_new['Name'] = names
iris_new.rename(columns={0:'SepalLength',1:'SepalWidth',2:'PetalLength', 3:'PetalWidth'}, inplace=True)
iris_new.head()
```

### 需求：画一个花瓣（petal）和花萼（sepal)长度的散点图，并且点的颜色要区分鸢尾花的种类

```python
color_map = dict(zip(iris.target_names, ['blue','green','red']))
```

```python
for species, group in iris_new.groupby('Name'):
    plt.scatter(group['PetalLength'], group['SepalLength'],
                color=color_map[species],
                alpha=0.3, edgecolor=None,
                label=species)
plt.legend(frameon=True, title='Name')
plt.xlabel('petalLength')
plt.ylabel('sepalLength')
```

## 用PetalLength和SepalLength作为横纵坐标作图

```python
sns.lmplot('PetalLength', 'SepalLength', iris_new, hue='Name', fit_reg=False)
```

## 用PetalWidth和SepalWidth作为横纵坐标作图

```python
sns.lmplot('PetalWidth', 'SepalWidth', iris_new, hue='Name', fit_reg=False)
```

```python
sns.pairplot(iris_new,hue="Name")
```

# 2. 趋势分析

```python
start = datetime(2015,9,20)
alibaba = pdr.get_data_yahoo('BABA', start=start)
amazon = pdr.get_data_yahoo('AMZN', start=start)
```

```python
#收盘价统计
alibaba['Adj Close'].plot(legend=True,figsize=(10,4))
```

```python
#成交量统计
alibaba['Volume'].plot(legend=True,figsize=(10,4))
```

## 增长幅度统计
#### 当天最高价 - 当天最低价

```python
alibaba['high-low'] = alibaba['High'] - alibaba['Low']
alibaba.head()
```

```python
alibaba['high-low'].plot(figsize=(10,4))
```

## 当天价格减去前一天价格的变化趋势

```python
alibaba['daily-return'] = alibaba['Adj Close'].pct_change() 
```

```python
alibaba['daily-return'].plot(figsize=(13,5),linestyle='--',marker='o')
```

### 直方图和密度度，从图中看出大多数变化集中在0附近，近似正态分布

```python
sns.distplot(alibaba['daily-return'].dropna(),bins=100,color='purple')
```

# 3. 风险分析

```python
start = datetime(2015,1,1)
company = ['AAPL','GOOG','MSFT','AMZN','FB']
top_tech_df = pdr.get_data_yahoo(company, start=start)['Adj Close']
top_tech_dr = top_tech_df.pct_change()
top_tech_dr.head()
```

```python
top_tech_df.plot()
```

```python
#比较谷歌和亚马逊股票的价格增长趋势
sns.jointplot('AMZN','GOOG', top_tech_dr, kind='scatter')
```

```python
#用散点图两两比较股票的价格增长趋势
sns.pairplot(top_tech_dr.dropna())
```

$$quantile() 计算公式为:\frac{\Delta S}{S} = \mu\Delta t + \sigma \epsilon \sqrt{\Delta t}$$ 

### 表示有95%的几率亏损幅度控制在2%范围内

```python
top_tech_dr.dropna()['AAPL'].quantile(0.05)
```

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

## 1. 数据分箱技术
#### 将数据分成不同的区间，并将每个区间的数据标记为不同的类型

```python
score_list = np.random.randint(25, 100, size=20)
```

```python
#分界点
bins = [0,59,70,80,100]
```

```python
#将25分到100分之间的分数按bins划分成不同的区间
score_cat = pd.cut(score_list, bins)
```

```python
#统计每个区间分数的数量
pd.value_counts(score_cat)
```

```python
df = DataFrame()
#随机生成学生姓名
df['student'] = [pd.util.testing.rands(3) for i in range(20)]
#得分
df['score'] = score_list
#得分程度，和bins区间相对应
df['Categories'] = pd.cut(df['score'],bins, labels=['Low','OK','Good','Great'])
df.head()
```

## 3. 数据聚合技术

```python
gr_new = df_gr.groupby(['color', 'country'])
gr_new.groups
```

```python
# 传入min聚合函数，为每列分组数据求最小值
gr_new.agg('min').head(2)
```

```python
# 传入max聚合函数，为每列分组数据求最小值
gr_new.agg('max').head(2)
```

```python
#自定义聚合函数
def foo(attr):
    return attr.max() - attr.min()
gr_new.agg(foo).head(2)
```

## 4. 透视表

```python
df_pivot  = pd.read_excel('sales-funnel.xlsx')
df_pivot.head()
```

```python
#统计每个销售销售情况，默认重复数据使用mean函数聚合
pd.pivot_table(df_pivot, index=['Name'],values=['Price','Quantity'])
```

```python
#使用sum聚合函数统计销售总体销售情况
pd.pivot_table(df_pivot, index=['Name'],values=['Price','Quantity'],aggfunc='sum')
```

```python
#使用双索引聚合，统计属于每个主管的销售员销售情况
#index：透视图索引，values：要透视的列，columns：在已知values的情况下透视其他列数据
pd.pivot_table(df_pivot, index=['Manager','Rep'],values=['Price','Quantity'],columns=['Product'])
```

```python
#增加columns[Product:产品]参数，这样values中的每个字段会对每种产品做统计
#针对Price字段，统计每种产品的价格
#针对Quantity字段，统计每种产品的数量
pd.pivot_table(df_pivot, index=['Manager','Rep'],values=['Price','Quantity'],columns=['Product'], fill_value=0, aggfunc='sum')
```

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

## 航班延时统计

```python
df = pd.read_csv('usa_flights.csv')
```

```python
df.shape
```

```python
df.head()
```

### 只统计晚点航班，arr_delay<0表示早点啦

```python
df['delayed'] = df['arr_delay'].apply(lambda x: x>0)
```

```python
df.head()
```

```python
group_by_carrier = df.groupby(['unique_carrier','delayed'])
```

```python
#获取AA航空公司，delayed=true的航班信息
group_by_carrier.get_group(('AA', True)).head()
```

```python
group_by_carrier.size()[:5]
```

## 用.unstack()函数将堆积数据转换成非堆积数据

![dataFrame.unstack()](unstack.png)

```python
%matplotlib inline
count_delays_by_carrier = group_by_carrier.size().unstack()
count_delays_by_carrier = count_delays_by_carrier.rename(index=str.upper, columns={False:'Undelayed',True:'Delayed'})
count_delays_by_carrier.head()
```

```python
count_delays_by_carrier.plot(kind='barh', stacked=True, figsize=[16,6], colormap='winter')
```
